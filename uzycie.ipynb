# Przypadek 1

```python
k = 1.2
a = 9.1
growthRate   = lambda m: k*np.sqrt(m)
devisionRate = lambda m: a*m

dt = 0.1
T  = 10.0

cells = SpaceOfCells(m = 1, dimX = 101, dimY = 101,
                     growthRate = growthRate, devisionRate = devisionRate,
                     dt = dt, T = T)

cells.growCells()
```

```python
cells.displayCells(6.0,"GrowthFrameNo1.svg", maxRadius = 10)
```

```python
cells.saveGrowthAnim("GrowthNo1.mp4",2,tMax = 2.0, maxRadius = 10)
```

```python
options = {
    'numBins'      : 40,
    'setCustomAxis': True,
    'xLeft'        : 0,
    'xRight'       : 1.1,
    'yUp'          : None,
    'yDown'        : None
}
cells.displayDistribution(nazwa='growHistAtTimeNo1.svg', t = 10.0, options = options)
```

```python
options = {
    'numBins'      : 40,
    'setCustomAxis': True,
    'xLeft'        : 0,
    'xRight'       : 1.1,
    'yUp'          : None,
    'yDown'        : None,
    'tMax'         : 6.0
}
cells.saveGrowhHistAnim(nazwa = "growHistNo1.mp4", interval = 2, options = options)
```
